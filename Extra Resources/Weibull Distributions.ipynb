```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Define the scale parameter (lambda)
scale = 1

# Define the shape parameters
shape_params = [1, 1.2, 1.5, 1.7, 2, 2.5, 3, 4, 5]

# Generate x values
x = np.linspace(0, 5, 1000)

# Plot Weibull distributions for each shape parameter
plt.figure(figsize=(10, 6))
for k in shape_params:
    y = weibull_min.pdf(x, k, scale=scale)
    plt.plot(x, y, label=f'k={k}')

# Add title and labels
plt.title('Weibull Distribution for Different k')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
