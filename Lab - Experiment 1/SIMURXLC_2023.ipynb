```python
# Defining the parameters with their default values
f = 50   # Frequency in Hz
Vrms = 240  # The phase rms Voltage in Volt
R = 8  # The Load Resistance in Ohm
XL = 10# The Load Inductive Reactance in Ohm
XC = 4  # The Load Capacitive Reactance in Ohm

# Displaying the parameters
print("The values used in the following calculations are: ")
print("The mains/supply Frequency f = {} Hz".format(f))
print("The phase rms Voltage Vrms = {} V".format(Vrms))
print("The Load Resistance R = {} Ohm".format(R))
print("The Load Inductive Reactance X_L = {} Ohm".format(XL))
print("The Load Capacitive Reactance X_C = {} Ohm".format(XC))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# time vector from 0 to 0.02 seconds
t = np.linspace(0, 0.02, 1000)

# define the angular frequency
omega = 2 * np.pi * f

# calculate the total reactance (X)
X = XL - XC

# calculate the total impedance (Z)
Z = np.sqrt(R**2 + X**2)

# calculate phase shift
phi = np.arctan(X/R)

# three phase voltages
Va = Vrms * np.sqrt(2) * np.sin(omega * t)
Vb = Vrms * np.sqrt(2) * np.sin(omega * t - 2*np.pi/3)
Vc = Vrms * np.sqrt(2) * np.sin(omega * t + 2*np.pi/3)

# three phase currents with phase shift
Ia = Vrms / Z * np.sin(omega * t - phi)
Ib = Vrms / Z * np.sin(omega * t - 2*np.pi/3 - phi)
Ic = Vrms / Z * np.sin(omega * t + 2*np.pi/3 - phi)
```

```python
# calculate instantaneous power for each phase
P1 = V1 * I1
P2 = V2 * I2
P3 = V3 * I3

# calculate total three-phase power
P_total = P1 + P2 + P3

# plotting powers with time
plt.figure(figsize=(10,6))
plt.plot(t, P1, label="Power Phase A", color='red')
plt.plot(t, P2, label="Power Phase B", color='green')
plt.plot(t, P3, label="Power Phase C", color='blue')
plt.plot(t, P_total, label="Three Phase Total Power")
plt.xlabel('Time (s)')
plt.ylabel('Power (W)')
plt.legend()
plt.title('Three Phase Instantaneous Powers')
plt.grid(True)
plt.show()
```
