A ferromagnetic core is shown in the following figure. The relative permeability of the core $\mu_r$ is 1000. The depth of the core is 5cm. The other dimensions are as shown in the figure below. 

i) Find the value of the current that will produce a flux $\phi$ of 0.005Wb.

ii) With this current, what is the flux density $B_{top}$ at the top of the core? 

iii) What is the flux at the right side $B_{right}$ of the core?



```python
from IPython.display import Image
Image(filename="Figures\TMC_Q2_Fig1.jpg",width=400,height=400)
```

### Answer

```python
N = 600 #[turns] The number of turns in the coil 
d = 0.05 # [m] depth of the core
mu_r = 1000 # relative permeability
wtop = 0.15 # [m] width of the top side
wbot = 0.15 # [m] width of the bottom side
wleft = 0.1 # [m] width of the left side
wright = 0.05 # [m] width of the right side
flux = 0.005 # [Wb] required flux in the core
c.mu_0 # Permeability of free space
print('mu_0 = {:.3e}'.format(c.mu_0))
```

There are three regions in the core. The top and bottom sections form one region ($l_1$), the left side forms a second region ($l_2$) and the right side forms a third region ($l_3$). 

If we assume that the mean path length of the flux is in the center of each leg of the core, and if we ignore spreading at the corners of the core, then the path lengths are:

```python
l1 = 2 * (0.2 + 0.1/2 + 0.05/2)
print('l1 = {:.2f} m'.format(l1))
```

```python
l2 = l3 =  0.15 + 0.15/2 + 0.15/2
print('l2 = l3 = {:.2f} m'.format(l2))
```

The reluctances of the regions are calculated from $$\mathcal{R} = \frac{l}{\mu_r \mu_0 A}$$

```python
Rel1 = l1 / (mu_r * c.mu_0 * d * wtop)
print('Rel1 = {:.2f} AT/Wb'.format(Rel1))
```

```python
Rel2 = l2 / (mu_r * c.mu_0 * d * wleft)
print('Rel2 = {:.2f} AT/Wb'.format(Rel2))
```

```python
Rel3 = l3 / (mu_r * c.mu_0 * d * wright)
print('Rel3 = {:.2f} AT/Wb'.format(Rel3))
```

The total reluctance is 

```python
Rtot = Rel1 + Rel2 + Rel3
print('Rtot = {:.2f} AT/Wb'.format(Rtot))
```

i) The MMF required to produce a flux of 0.005 Wb is $MMF = \phi \cdot R_{total}$

```python
MMF = flux * Rtot
print('MMF = {:.2f} AT'.format(MMF))
```

The required current is $I = \dfrac{MMF}{N}$

```python
I = MMF/N
print('I = {:.2f} A'.format(I))
```

ii) Flux Density can be found from the flux and the cross-sectional area as $B = \dfrac{\phi}{A}$

The flux density on the top of the core is 

```python
Btop = flux / (d * wtop)
print('Btop = {:.2f} AT'.format(Btop))
```

The flux density on the right side of the core is 

```python
Bright = flux / (d * wright)
print('Bright = {:.2f} AT'.format(Bright))
```
