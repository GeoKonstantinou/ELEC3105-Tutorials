A transformer core with an effective mean path length of $10 \ in$ has a 300-turn coil wrapped around one leg. Its cross-sectional area is $0.25\ in^2$. Its magnetization curve is shown in the Figure below. If a current of $0.3\,A$ is flowing in the coil.
 * What is the total flux in the core? 
 * What is the flux density?

```python
from IPython.display import Image
Image(filename="Figures\TMC_Q8_Fig1.png",width=600,height=600)
```

### Answer

```python
c = 0.0254 # Conversion constant [m/in]
lc = 10 * c # [m]
A = 0.25 * c**2 # [m²]
N = 300 # [turns]
I = 0.3  # [A]
```

The magnetizing intensity applied to this core is:

$$H = \frac{\mathcal{F}}{l_c} = \frac{Ni}{l_c}$$

```python
H = N*I / lc
print('H = {:.2f} At/m'.format(H))
```

From the magnetization curve, the flux density in the core is:

```python
B = 1.35  # [T]
```

The total flux in the core is $\phi = B\cdot A$

```python
phi = B*A
print('ϕ = {:.3f} mWb'.format(phi*1000))
```
