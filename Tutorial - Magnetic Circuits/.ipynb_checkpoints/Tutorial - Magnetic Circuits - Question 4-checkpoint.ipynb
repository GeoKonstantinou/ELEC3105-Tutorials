A two-legged core is shown in the following figure. The winding on the left leg of the core ($N_1$) has 400 turns and the winding on the right  ($N_2$) has 300 turns. The coils are would in the directions shown in the figure. If the dimensions are as show, what is the flux produced by currents $i_1$ = 0.5A and $i_2$ = 0.75A. 

Assume that the relative permeability of the core is $\mu_r$ = 1000 and constant. 

```python
from IPython.display import Image
Image(filename="Figures\TMC_Q4_Fig1.jpg",width=500,height=500)
```

### Answer

```python
N1 = 400 #[turns] The number of turns in coil 1
N2 = 300 #[turns] The number of turns in coil 2
d = 0.15 # [m] depth of the core
mu_r = 1000 # relative permeability
htop = 0.15 # [m] height of the top side
hmid = 0.5 # [m] height of the middle 
hbot = 0.15 # [m] height of the bottom side
lleft = 0.15 # [m] length of the left side
lright = 0.15 # [m] length of the right side
lmid = 0.5 # [m] length of the middle 
width = 0.15 # [m] average width of core
I1 = 0.5 # [A] Current on coil 1
I2 = 0.75 # [A] Current on coil 2
c.mu_0 # Permeability of free space
```

### Answer

The two coils are wound so that their magnetomotive forces are added, so that the total MMF on the core is $$\mathcal{F_{Tot}} = N_1I_1 + N_2 I_2$$

```python
MMF = N1 * I1 + N2 * I2
print('MMF = {:.2f} AT'.format(MMF))
```

The total reluctance of the core is $\mathcal{R} = \dfrac{l}{\mu_r \mu_0 A}$

```python
Rel = (2*hmid + 2*lmid + 2* lleft/2 + 2*lright/2 + 2*htop/2 + 2*hbot/2)/(mu_r * c.mu_0 * d * width)
print('Rel = {:.2f} AT/m2'.format(Rel))
```

The flux of the core is $\phi = \frac{\mathcal{F_{Tot}}}{\mathcal{R_{Tot}}}$

```python
flux  = MMF/Rel
print(' = {:.2f} mWb'.format(flux*1000))
```
