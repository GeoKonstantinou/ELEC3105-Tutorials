For the magnetic circuit of the following figure, the number of turns N is 10, the length of the air-gap is 0.1mm and the average length of the core, which is made out of M-19 material, is 100mm. Also consider a stacking factor of 0.9. What is the current required to establish a 1T flux density in the airgap?

```python
from IPython.display import Image
Image(filename="Figures\TMC_Q14_Fig1.png",width=500,height=500)
```

### Answer

```python
N = 10 # [turns] Turns of the coil
lg = 0.0001 # [m] length of the airgap
lc = 0.1 # [m] length of the core
Bg = 1 # [T] magnetic flux required in the airgap
SF = 0.9 # Stacking Factor
```

```python
Hg = Bg / c.mu_0
print('Hg = {:.6f} A/m'.format(Hg))
```

Using Ampere's Law the MMF can be calculated as $MMF = H_g \cdot l_g$

```python
MMF_g = Hg * lg
print('MMFg = {:.2f} AT'.format(MMF_g))
```

Considering the stacking factor for the core, we find the required magnetic flux density inside the core as $B_c = \dfrac{B_g}{SF}$

```python
Bc = Bg / SF
print('Bc = {:.2f} T'.format(Bc))
```

Using the magnetisation curve for the core we can find the magnetic field intensity as 130AT/m and the corresponding MMF

```python
Hc = 130 # [AT/m] magnetic field intensity in the core
MMF_c = Hc * lc
print('MMFc = {:.2f} AT'.format(MMF_c))
```

```python
MMF_total = MMF_c + MMF_g
print('MMF = {:.2f} AT'.format(MMF_total))
```

The required current to create this MMF in the core is then $I = \dfrac{MMF}{N}$

```python
I = MMF_total/N
print('I = {:.2f} A'.format(I))
```
